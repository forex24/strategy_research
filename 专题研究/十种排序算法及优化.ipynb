# 排序算法

## 冒泡排序

```python
def bubble_sort(arr):
    arr_length = len(arr)
    for i in range(arr_length-1): # 每每趟的截至点不断左移
        to_i = arr_length-1-i 
        for j in range(to_i):  # 从左到右顺序比较+交换
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

## 冒泡排序优化（某趟没有交换提前终止）

```python
def bubble_sort_opt(arr):
    arr_length = len(arr)
    for i in range(arr_length-1): # 每趟的截至点不断左移
        to_i = arr_length-1-i 
        is_ex = False
        for j in range(to_i):  # 从左到右顺序比较+交换
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                is_ex = True
        if not is_ex: # 没有交换过，全部数据已经有序，可以提前跳出
            break
    return arr
```

## 选择排序

```python
def quick_sort(arr):
    arr_len = len(arr)
    # 共arr_len-1趟，每次起始点向后平移1
    for i in range(arr_len): 
        # 用起始点数值初始化最小值
        min_v = arr[i] 
        min_i = i
        # 顺序向右扫描得到真正的最小值
        for j in range(i+1, arr_len):
            if arr[j]<min_v:
                min_v = arr[j]
                min_i = j
        arr[i], arr[min_i] = arr[min_i], arr[i] # 交换真正最小值到第i位
    return arr            
```

## 插入排序

```python
def insert_sort(arr):
    arr_len = len(arr)
    # 一共循环 arr_len-1 次
    for i in range(1, arr_len):
        v_i = arr[i]
        # 从右往左与v_i进行比较
        for j in range(i, -1, -1):
            if arr[j-1]>v_i: # 若arr[j-1]大于v_i，则arr[j-1]右移1，继续向左搜索
                arr[j] = arr[j-1]
            else:          # 反之，则无需再搜索，跳出循环
                break
        arr[j] = v_i  # v_i插入到arr[j]的位置
    return arr
```

## 希尔排序

```python
def shell_sort(arr):
    arr_len = len(arr)
    gap = arr_len // 2
    while gap > 0:
        for i in range(gap, arr_len):
            while i>=gap and arr[i - gap] > arr[i]:
                arr[i - gap], arr[i] = arr[i], arr[i - gap]
                i -= gap
        gap //= 2
    return arr
```

## 快速排序

```python
def quick_sort(arr, start_i, end_i):
    # 递归退出条件
    if start_i>=end_i:
        return 

    mid_v = arr[start_i] # 基准
    low_i = start_i
    high_i = end_i
    while low_i<high_i:
        while low_i<high_i and arr[high_i]>=mid_v: # 从右往左搜索大于等于mid_v的数
            high_i -= 1
        arr[low_i] = arr[high_i] # 填坑
        while low_i<high_i and arr[low_i]<mid_v:  # 从左往右搜索小于mid_v的数
            low_i += 1
        arr[high_i] = arr[low_i]

    # low_i 与 high_i 重合，退出循环
    arr[low_i] = mid_v
    
    quick_sort(arr, start_i, low_i-1)
    quick_sort(arr, low_i+1, end_i)
    
    return arr
```

## 归并排序

```python
def merge_sort(arr, l, r):
    """ 
        递归进行归并排序
            l：左界index，迭代第一层时为0
            r：右界index，迭代第一层时为len(arr)-1
    """
    #  递归退出条件：分组到仅剩一个元素（l==r)
    if r > l:
        # 分成两组并行排序
        m = (l+r)//2   
        merge_sort(arr, l, m) 
        merge_sort(arr, m, r) 
        
        # 归总
        merge(arr, l, m, r) 
    return arr
        
def merge(arr, l, m, r):
    """ 
        通过构建一个临时数组，对 [l, m], [m+1, r] 两组（已排好序的）子数组进行迭代排序
            arr： 原始的数组
            l：   左界
            m：   中界
            r：   右界
    """
    L = arr[l: m+1].copy()  # 长度为 【1, arr_len//2+1】
    R = arr[m+1: r+1].copy()  # 长度为 【1, arr_len//2+1】
    
    L_i, L_len = 0, m-l+1
    R_i, R_len = 0, r-m+1
    
    arr_i = l
    while L_i< L_len and R_i<R_len:
        L_v = L[L_i]
        R_v = R[R_i]
        if L_v<=R_v:
            arr[arr_i] = L_v
            L_i += 1
        else:
            arr[arr_i] = R_v
            R_i += 1            
        arr_i += 1
    
    while L_i< L_len:
        L_v = L[L_i]
        arr[arr_i] = L_v
        L_i += 1
        arr_i += 1

    while R_i<R_len:
        R_v = R[R_i]
        arr[arr_i: r+1] = R_v
        R_i += 1
        arr_i += 1
```

## 堆排序

```python
def heapify(arr, n, i):
    """ 
        arr：存储堆的数组
        n：  数组的长度
        i：  待维护节点的下标
    """
    # 找到子节点的index
    lson_i = 2 * i + 1
    rson_i = 2 * i + 2

    # 初始化largest_i，跟左右子节点进行比较，确定最大值
    largest_i = i
    if lson_i<n and arr[lson_i]>arr[largest_i]:
        largest_i = lson_i
    if rson_i<n and arr[rson_i]>arr[largest_i]:
        largest_i = rson_i
    
    # 最大值为子节点，需要交换
    if largest_i!=i:  
        arr[largest_i], arr[i] = arr[i], arr[largest_i]
        heapify(arr, n, largest_i) # 对子节点递归进行heapify
    return 
    
def heap_sort(arr):
    # 1. 建大顶堆 (从【最后一个叶节点的父节点】，一路向上到根调用heapify)
    n = len(arr)
    last_leaf_f_i = (n-1)//2
    for i in range(last_leaf_f_i, -1, -1):
        heapify(arr, n, i)
    
    # 2. 排序
    
    return arr
```
